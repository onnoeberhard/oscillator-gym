```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
Qs = [1, 2, 5, 10, 20]
fs = [1/10, 1/5, 1/2, 1, 2, 5, 10]
N = 50

results = np.load('../tmp/oscillator_analysis.npy').sum(0) / N
```

```python
fig, ax = plt.subplots(constrained_layout=True)

im = ax.pcolormesh(results, cmap='viridis', rasterized=True)
ax.set_title("White noise agent on Oscillator-v0")

for x in range(len(Qs)):
    for y in range(len(fs)):
        t = results[x, y]
        ax.text(y + 0.5, x + 0.5, f"{t:.2f}", ha='center', va='center', color='w' if t < 0.5 else 'k')

ax.set_xlabel("Natural frequency $f$")
ax.set_xticks(np.arange(len(fs)) + 0.5)
ax.set_xticklabels(fs)
ax.set_ylabel("Quality factor $Q$")
ax.set_yticks(np.arange(len(Qs)) + 0.5)
ax.set_yticklabels(Qs)
fig.colorbar(im, ax=ax, label="Success rate")
```
